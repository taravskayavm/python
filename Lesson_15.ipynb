```python
import random
import pandas as pd
```

```python
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI' : lst})
data.head()
```

```python
#Создаем пустой DataFrame с нужными столбцами
one_hot_data = pd.DataFrame(columns=['robot', 'human'])

#Заполняем one-hot DataFrame
for category in data['whoAmI'].unique():
  one_hot_data[category] = (data['whoAmI'] == category).astype(int)

one_hot_data.head()
```
